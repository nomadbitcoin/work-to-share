```python
class facebookInsights():
    def driver_(self):
        options = webdriver.ChromeOptions()
        options.add_argument("profile")
        options.add_argument("user-data-dir=Browser/Profile")

        self.driver = webdriver.Chrome(executable_path='../Browser/Driver/chromedriver', options=options)
        
    def typeCity(self, city):
        ''' Recebe uma cidade e digita na busca'''
        try:
            city_input = self.driver.find_element_by_xpath('//input[@placeholder="País, região ou cidade"]')
            city_input.click()
            city_input.send_keys(city)
            time.sleep(1)
            city_input.send_keys(Keys.ENTER)
            return True
        except Exception as error:
            if 'ElementClickInterceptedException' in str(error):
                input('Aperte para continuar')
                return True
```
